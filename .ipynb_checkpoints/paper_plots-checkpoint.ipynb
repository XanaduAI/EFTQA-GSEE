```python
def estimate_samples(eta, tau, eps, theta = 0.05):
    gamma = 0.57721567
    D = np.sqrt(2)/(tau*eps)*np.log(4*np.sqrt(2*np.pi/(tau*eps))*(2+tau/eps))
    if eps>eta/2:
        return None
    A = (2.07/np.pi*(np.log(4*D) + gamma)+1)
    M_ = (A/(eta/2-eps))**2 *(np.log(np.log(1/(tau*eps))) + np.log(1/theta))
    return M_

tau =  np.load('results/{}/tau.npy'.format(31))
fontsize=16

epsilon = [0.00001,0.0001,0.001,0.01,0.1]
colors = ['darkorange','red','darkviolet','blue', 'darkgreen']
eta = np.logspace(-5,0,100)
# print(eta)

plt.figure()

for _,eps in enumerate(epsilon):
    M = [estimate_samples(e, tau,eps, theta = 0.05) for e in eta]
    if eps ==0.1:
        plt.plot(M,eta,label='$ \epsilon = 10^{{-1}}{}$',color=colors[_])
    elif eps ==0.01:
        plt.plot(M,eta,label='$\epsilon = 10^{{-2}}{}$',color=colors[_])
    elif eps ==0.001:
        plt.plot(M,eta, label='$\epsilon = 10^{{-3}}{}$',color=colors[_])
    elif eps ==0.0001:
        plt.plot(M,eta,label='$\epsilon = 10^{{-4}}{}$',color=colors[_])
    elif eps ==0.00001:
        plt.plot(M,eta,label='$\epsilon = 10^{{-5}}{}$',color=colors[_])
    
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'resolution $\eta$', fontsize = fontsize)
plt.xlabel(r'number of samples $M$',fontsize=fontsize)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.savefig('measurements.pdf',dpi=800)    
```

```python
def find_rpt(acdf1):
   
    l = 500
    
    
    sigma = np.std(acdf1[:l])
    # print(acdf1[:l])
    epsilon =1*np.mean(np.absolute(acdf1[:l]))#+0.001
    # epsilon =  min(0.001,epsilon)
    # print(sigma,epsilon)
    
    acdf1 = acdf[l:]
    # print(sigma)
    cps = [len(acdf1)]
    alpha = 1-0.15
    # print(sigma)
    for i in range(1000):

        acdf_trial = acdf1[:cps[-1]]
        # plt.figure()
        # plt.plot(acdf_trial)
        # plt.show()
        
        algo = rpt.KernelCPD().fit(acdf_trial)
        trial = algo.predict(1)[0]
       

        F = check_stat_relevant(acdf_trial,trial)
    #     print(F,trial)
        cps.append(trial)
        df = len(acdf_trial)-2
        p_value = scipy.stats.f.cdf(F, 1, df)
        # print(sigma,p_value, acdf_trial[trial-10:trial].mean())
        
        if p_value < alpha:
            print('p value', p_value, alpha)
            break
        if acdf_trial[trial-10:trial].mean() < 1*sigma+epsilon:
            
            # print('break',acdf_trial[trial-20:trial].mean(), 2*sigma+epsilon)
            break
        if trial<10:
            break
        
    
        
    return np.array(cps)+l
```

```python
initial_state = 'dmrg_10_26_-1_1'
wf = np.load('data/{}.npy'.format(initial_state)).reshape(-1)

# wf = np.random.uniform(-1,1,2**3)
# wf = wf/np.linalg.norm(wf)

print(np.linalg.norm(wf,ord=2))

def keep_largest(vector, l):
    sorted_indices = np.argsort(vector)[::-1]
    threshold_index = min(len(vector), l) - 1
    threshold_value = vector[sorted_indices[threshold_index]]
    wf = np.where(vector >= threshold_value, vector, 0)

    return wf/np.linalg.norm(wf)


trunc = np.arange(1,101)

states = [keep_largest(wf,l) for l in trunc]

# print(wf)
# print(keep_largest(wf,9))
# print(np.dot(wf.conjugate().transpose(),keep_largest(wf,1)))
# print(np.linalg.norm((wf - keep_largest(wf,9))))

norm = [np.linalg.norm(wf-keep_largest(wf,l)) for l in trunc]
fid = [np.dot(wf.conjugate().transpose(),s) for s in states]
```
